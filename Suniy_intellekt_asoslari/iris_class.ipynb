```python
import numpy as np
x=np.loadtxt('D:\Projects\Algo_daily\iris.data',usecols=(0,1,2,3),delimiter=',')
# print(x)

def kv(x,y):
    b = 0
    for i in range(len(x)):
        b += (x[i]-y[i])**2
    return b**(1/2)

b = np.zeros(shape=(150,150))
for i in range(len(x)):    
    for j in range(i,len(x)):   
        b[i][j] = kv(x[i],x[j])

# print(b)

maksimal = 0
i1 = 0
j1 = 0
for i in range(len(b)):
    for j in range(i,len(b)):
        if b[i][j] > maksimal:
            maksimal = b[i][j]
            i1 = i
            j1 = j

print(f"eng uzoq masofa: {maksimal} qator: {i1} ustun: {j1}")

minimal = maksimal
i2 = 0
j2 = 0
for i in range(len(b)):
    for j in range(i,len(b)):
        if b[i][j] < minimal and b[i][j] != 0:
            minimal = b[i][j]
            i2 = i
            j2 = j

print(f"eng yaqin masofa: {minimal} qator: {i2} ustun: {j2}")
```

### 2)Quyidagi funksiyalar berilgan: **y1=2x+1,  y2=−x+7**
 a)x ning 0 dan 5 gacha qiymatlari uchun:
 y₁ va y₂ ni hisoblang

```python
def hisobla():
    for x in range(6):
        y1 = 2*x+1
        y2 = -x+7
        print(f'x={x} ---> y1={y1} *** y2={y2}')
hisobla()
```

b)Har bir x uchun:
qaysi funksiya katta ekanini aniqlang


```python
def hisobla():
    for x in range(6):
        y1 = 2*x+1
        y2 = -x+7
        if y1>y2:
            print(f'x={x} ---> y1={y1} *** y2={y2}*********Kattasi{y1}.')
        elif y2>y1:
            print(f'x={x} ---> y1={y1} *** y2={y2}*********Kattasi{y2}.')
        else:
            print(f'x={x} ---> y1={y1} *** y2={y2}*********Ikkalasi teng.')
        
hisobla()
```

c)Nechta holatda y₁ > y₂ bo‘lganini hisoblang


```python
def hisobla():
    count = 0
    for x in range(6):
        y1 = 2*x+1
        y2 = -x+7
        if y1>y2:
            count+=1
    return count
hisobla()
```

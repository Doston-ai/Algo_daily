```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random

def euclidean(point, data):
    data = np.array(data)
    return np.sqrt(np.sum((point - data)**2, axis=1))

class KMeans:
    def __init__(self, n_clusters=8, max_iter=1):
        self.n_clusters = n_clusters
        self.max_iter = max_iter

    def fit(self, X):
        self.markaz = [random.choice(X)]

        for _ in range(self.n_clusters - 1):
            masofa = np.sum([euclidean(markaz, X) for markaz in self.markaz], axis=0)
            masofa /= np.sum(masofa)
            yangi_markaz_idx, = np.random.choice(range(len(X)), size=1, p=masofa)
            self.markaz += [X[yangi_markaz_idx]]

        iteration = 0
        oldingi_markazlar = None
        while np.not_equal(self.markaz, oldingi_markazlar).any() and iteration < self.max_iter:

            # ✅ self.guruhlar — har cluster ning nuqtalari saqlanadi
            self.guruhlar = [[] for _ in range(self.n_clusters)]
            for a in X:
                masofa = euclidean(a, self.markaz)
                markaz_idx = np.argmin(masofa)
                self.guruhlar[markaz_idx].append(a)

            oldingi_markazlar = self.markaz
            self.markaz = [np.mean(clusters, axis=0) for clusters in self.guruhlar]

            for i, markaz in enumerate(self.markaz):
                if np.isnan(markaz).any():
                    self.markaz[i] = oldingi_markazlar[i]

            iteration += 1

    def evaluate(self, X):
        markazlar = []
        markazlar_idxs = []
        for a in X:
            masofa = euclidean(a, self.markaz)
            markaz_idx = np.argmin(masofa)
            markazlar.append(self.markaz[markaz_idx])
            markazlar_idxs.append(markaz_idx)
        return markazlar, markazlar_idxs

    def plot(self):
        k = self.n_clusters
        colors = ['r', 'b', 'g', 'y', 'purple', 'orange', 'pink', 'cyan']

        plt.figure(figsize=(10, 10))

        for i, guruh in enumerate(self.guruhlar):
            xs = [p[0] for p in guruh]
            ys = [p[1] for p in guruh]
            plt.scatter(xs, ys, color=colors[i % len(colors)], label=f'Guruh {i+1}')

        for c in self.markaz:
            plt.scatter(c[0], c[1], color='k', s=200, marker='X')

        plt.title(f'K-means (k = {k})')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.legend()
        plt.grid(alpha=0.4)
        plt.show()


# ── Ishlatish ─────────────────────────────────────────────
X = pd.read_csv(r'D:\Projects\Algo_daily\kmeans_2000_points.csv')
X = X.values

random.seed(42)
kmeans = KMeans(n_clusters=4, max_iter=10)
kmeans.fit(X)
# chizma
kmeans.plot()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import pandas as pd
import random

# ============================================================
# EVKLID MASOFASI
# ============================================================
def euclidean(point, data):
    data = np.array(data)
    return np.sqrt(np.sum((point - data)**2, axis=1))


# ============================================================
# KMEANS KLASSI — tuzatilgan
# ============================================================
class KMeans:
    def __init__(self, n_clusters=8, max_iter=100):
        self.n_clusters = n_clusters
        self.max_iter = max_iter

    def fit(self, X):
        self.markaz = [random.choice(X)]

        for _ in range(self.n_clusters - 1):
            masofa = np.sum([euclidean(markaz, X) for markaz in self.markaz], axis=0)
            masofa /= np.sum(masofa)
            yangi_markaz_idx, = np.random.choice(range(len(X)), size=1, p=masofa)
            self.markaz += [X[yangi_markaz_idx]]

        iteration = 0
        oldingi_markazlar = None
        while np.not_equal(self.markaz, oldingi_markazlar).any() and iteration < self.max_iter:
            sorted_points = [[] for _ in range(self.n_clusters)]
            for a in X:
                masofa = euclidean(a, self.markaz)
                markaz_idx = np.argmin(masofa)
                sorted_points[markaz_idx].append(a)

            oldingi_markazlar = self.markaz
            self.markaz = [np.mean(clusters, axis=0) for clusters in sorted_points]

            # ✅ NaN tekshiruvi — bo'sh cluster bo'lsa eskisini saqla
            for i, markaz in enumerate(self.markaz):
                if np.isnan(markaz).any():
                    self.markaz[i] = oldingi_markazlar[i]

            iteration += 1

    def evaluate(self, X):
        markazlar = []
        markazlar_idxs = []
        for a in X:
            masofa = euclidean(a, self.markaz)   # ✅ self.markaz (self.markazlar emas)
            markaz_idx = np.argmin(masofa)
            markazlar.append(self.markaz[markaz_idx])
            markazlar_idxs.append(markaz_idx)
        return markazlar, markazlar_idxs


# ============================================================
# DATA YUKLASH
# ============================================================
X = pd.read_csv(r'D:\Projects\Algo_daily\kmeans_2000_points.csv')
X = X.values

N_CLUSTERS = 6

# ============================================================
# KMEANS ISHLATISH
# ============================================================
random.seed(42)
kmeans = KMeans(n_clusters=N_CLUSTERS)
kmeans.fit(X)
markazlar, labels = kmeans.evaluate(X)

centroids_array = np.array(kmeans.markaz)
labels_array    = np.array(labels)


# ============================================================
# VIZUALIZATSIYA
# ============================================================
COLORS = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4',
          '#FFEAA7', '#DDA0DD', '#98D8C8', '#F7DC6F']

fig, axes = plt.subplots(1, 2, figsize=(16, 7))
fig.patch.set_facecolor('#0D1117')

# ── Chap: Nuqtalar + Markazlar ─────────────────────────────
ax1 = axes[0]
ax1.set_facecolor('#161B22')

for i in range(N_CLUSTERS):
    mask = labels_array == i
    ax1.scatter(X[mask, 0], X[mask, 1],
                c=COLORS[i], alpha=0.65, s=45,
                edgecolors='white', linewidths=0.3,
                label=f'Cluster {i+1}  ({mask.sum()} ta)')

# Markazlar
ax1.scatter(centroids_array[:, 0], centroids_array[:, 1],
            c='white', marker='*', s=400,
            edgecolors='black', linewidths=1.2,
            zorder=5, label='Markazlar')

# Markaz raqamlari
for i, (cx, cy) in enumerate(centroids_array):
    ax1.annotate(f'  C{i+1}', (cx, cy),
                 color='white', fontsize=10, fontweight='bold',
                 bbox=dict(boxstyle='round,pad=0.25',
                           facecolor=COLORS[i], alpha=0.75))

ax1.set_title('KMeans — Nuqtalar va Markazlar',
              color='white', fontsize=13, fontweight='bold', pad=12)
ax1.set_xlabel('X', color='#8B949E', fontsize=11)
ax1.set_ylabel('Y', color='#8B949E', fontsize=11)
ax1.tick_params(colors='#8B949E')
for spine in ax1.spines.values():
    spine.set_edgecolor('#30363D')
ax1.legend(facecolor='#21262D', edgecolor='#30363D',
           labelcolor='white', fontsize=9, loc='best')

# ── O'ng: Voronoi chegaralari ──────────────────────────────
ax2 = axes[1]
ax2.set_facecolor('#161B22')

# Voronoi fon
xx, yy = np.meshgrid(
    np.linspace(X[:, 0].min() - 1, X[:, 0].max() + 1, 400),
    np.linspace(X[:, 1].min() - 1, X[:, 1].max() + 1, 400)
)
grid_points = np.c_[xx.ravel(), yy.ravel()]
_, grid_labels = kmeans.evaluate(grid_points)
grid_labels = np.array(grid_labels).reshape(xx.shape)

cmap = ListedColormap([COLORS[i] for i in range(N_CLUSTERS)])
ax2.contourf(xx, yy, grid_labels, alpha=0.15, cmap=cmap)
ax2.contour(xx, yy, grid_labels, colors='white', alpha=0.25, linewidths=1)

# Nuqtalar
for i in range(N_CLUSTERS):
    mask = labels_array == i
    ax2.scatter(X[mask, 0], X[mask, 1],
                c=COLORS[i], alpha=0.65, s=45,
                edgecolors='white', linewidths=0.3)

# Markazdan nuqtalarga ingichka chiziq
for i in range(N_CLUSTERS):
    mask = labels_array == i
    for pt in X[mask]:
        ax2.plot([centroids_array[i, 0], pt[0]],
                 [centroids_array[i, 1], pt[1]],
                 color=COLORS[i], alpha=0.08, linewidth=0.5)

# Markazlar
ax2.scatter(centroids_array[:, 0], centroids_array[:, 1],
            c='white', marker='*', s=400,
            edgecolors='black', linewidths=1.2, zorder=6)

for i, (cx, cy) in enumerate(centroids_array):
    ax2.annotate(f'C{i+1}\n({cx:.1f}, {cy:.1f})', (cx, cy),
                 textcoords="offset points", xytext=(8, 8),
                 color='white', fontsize=8, fontweight='bold',
                 bbox=dict(boxstyle='round,pad=0.3',
                           facecolor=COLORS[i], alpha=0.75))

ax2.set_title(f'Voronoi Chegaralari  (k={N_CLUSTERS})',
              color='white', fontsize=13, fontweight='bold', pad=12)
ax2.set_xlabel('X', color='#8B949E', fontsize=11)
ax2.set_ylabel('Y', color='#8B949E', fontsize=11)
ax2.tick_params(colors='#8B949E')
for spine in ax2.spines.values():
    spine.set_edgecolor('#30363D')

# ── Sarlavha ───────────────────────────────────────────────
fig.suptitle(f'KMeans Cluster Tahlili  (k={N_CLUSTERS})',
             color='white', fontsize=16, fontweight='bold', y=1.01)

plt.tight_layout()
plt.savefig('kmeans_result.png', dpi=150,
            bbox_inches='tight', facecolor='#0D1117')
plt.show()

# ============================================================
# STATISTIKA
# ============================================================
print("\n📊 CLUSTER STATISTIKASI:")
print("─" * 45)
for i in range(N_CLUSTERS):
    mask = labels_array == i
    cx, cy = centroids_array[i]
    print(f"Cluster {i+1}: {mask.sum():5d} ta nuqta  |  "
          f"Markaz → ({cx:.2f}, {cy:.2f})")
print("─" * 45)
print(f"Jami: {len(X)} ta nuqta, {N_CLUSTERS} ta cluster")
```

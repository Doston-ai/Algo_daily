```python
# o'rta qiymat (mean)
x = [2, 4, 6, 8, 10]
y = sum(x) / len(x)
print("O'rta qiymat:", y)
```

```python
# median
x = [2, 4, 6, 8, 10]
x.sort()
n = len(x)
if n % 2 == 0:
    median = (x[n//2 - 1] + x[n//2]) / 2
else:
    median = x[n//2]
print("Median:", median)
```

```python
# mode
data = [2, 4, 6, 8, 8, 10, 10] 
takror = {}

for i in data:
    if i not in takror:
        takror[i] = 1
    else:
        takror[i] += 1
max_count = max(takror.values())
# mode = [key for key, value in takror.items()  if value == max_count] # one line code
mode = []
for key, value in takror.items():
    if value == max_count:
        mode.append(key)

print("Lug'at >>>", takror)       
print("Max takrorlangan >>>", max_count) 
print("Moda >>>", mode) 
```

```python
# dispersion    
x = [2, 4, 8, 6, 8, 10]
x.sort()
mean = sum(x) / len(x)
variance = sum((xi - mean) ** 2 for xi in x) / len(x)
print("Dispersion (Variance):", variance)
```

```python
def get_minor(matrix, row, col):
    """Berilgan qator va ustunni o'chirib minor matrits qaytaradi"""
    return [
        [matrix[i][j] for j in range(len(matrix)) if j != col]
        for i in range(len(matrix)) if i != row
    ]

def determinant(matrix):
    n = len(matrix)

    # Bazis holat: 1x1
    if n == 1:
        return matrix[0][0]

    # Bazis holat: 2x2
    if n == 2:
        return matrix[0][0] * matrix[1][1] - matrix[0][1] * matrix[1][0]

    det = 0
    for j in range(n):
        # Kofaktor: (-1)^(0+j) * minor determinanti
        sign = (-1) ** j
        det += sign * matrix[0][j] * determinant(get_minor(matrix, 0, j))

    return det


# --- Testlar ---
a2 = [[3, 2],
      [1, 4]]

a3 = [[1, 2, 3],
      [4, 5, 6],
      [7, 8, 10]]

a4 = [[2, 3, 1, 5],
      [1, 0, 3, 1],
      [0, 2, -3, 2],
      [0, 2, 3, 1]]

print(f"2x2 det = {determinant(a2)}")
print(f"3x3 det = {determinant(a3)}")
print(f"4x4 det = {determinant(a4)}")


```
